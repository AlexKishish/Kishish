```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from datetime import datetime
from matplotlib import rc

# Установка русского шрифта
rc('font', family='Arial')

# Загрузка данных из файла
file_path = 'time_messagees.txt'
data = pd.read_csv(file_path, header=None, names=['Time', 'Messages'], delimiter=',')

# Преобразование времени в секунды
data['Time'] = data['Time'].apply(lambda x: datetime.strptime(x, '%H:%M:%S').second)

# Построение модели линейной регрессии
X = data[['Time']]
y = data['Messages']
model = LinearRegression()
model.fit(X, y)

# Предсказание значений
data['Predictions'] = model.predict(X)

# Оценка модели
r_squared = model.score(X, y)
print(f'R-squared: {r_squared}')

# Построение графика
plt.scatter(data['Time'], data['Messages'], label='Фактические данные')
plt.plot(data['Time'], data['Predictions'], color='red', label='Модель линейной регрессии')
plt.xlabel('Время (секунды)')
plt.ylabel('Количество сообщений')
plt.title('Модель линейной регрессии для данных мобильного оператора')
plt.legend()
plt.show()
```
